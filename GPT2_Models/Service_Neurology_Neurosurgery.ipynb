```
import pandas as pd
Service_Hospital = pd.read_csv('/content/drive/My Drive/Data-X Project/Colab Notebooks/Service_Brief_Hospital_Course_Top5.csv')
Service_Hospital.shape
```

```
Service_Hospital['Service'].value_counts()
```

```
Service_NEUROLOGY = Service_Hospital[Service_Hospital['Service']=='NEUROLOGY' ].copy()
```

```
Service_NEUROSURGERY = Service_Hospital[Service_Hospital['Service']=='NEUROSURGERY' ].copy()
```

```
Service_NEURO= pd.concat([Service_NEUROLOGY,Service_NEUROSURGERY], axis=0)
```

```
Service_NEURO.head(3)
```

```
Service_NEURO=Service_NEURO['Brief Hospital Course'].copy()
```

```
Service_NEURO.shape
```

```
Service_NEURO_train= Service_NEURO.sample(frac=0.85, random_state=0)
Service_NEURO_test = Service_NEURO.drop(Service_NEURO_train.index)
```

```

Service_NEURO_test.to_csv(r'/content/drive/My Drive/Data-X Project/Colab Notebooks/Data/Service_NEURO_test.txt', header=None, index=None, sep='\t', mode='a')
Service_NEURO_train.to_csv(r'/content/drive/My Drive/Data-X Project/Colab Notebooks/Data/Service_NEURO_train.txt', header=None, index=None, sep='\t', mode='a')
```
